```python
import torch
import poptorch

x = torch.randn(2**8, 2**10)
w = torch.randn(2**10, 2**12)

class MatrixMul(torch.nn.Module):
    def forward(self, x, w):
        y = x @ w * 2**-5
        z = y.std()
        return z

model = MatrixMul()

# Wrap the model for inference on the IPU
poptorch_model = poptorch.inferenceModel(model)

# Run the model on the IPU and print the result
result = poptorch_model(x, w)
print(result)
```
