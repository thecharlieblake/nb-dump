```python
import torch
import poptorch

x = torch.randn(2**10, 2**12)
w = torch.randn(2**12, 2**14)

class MatrixMul(torch.nn.Module):
    def forward(self, x, w):
        y = x @ w * 2**-6
        z = y.std()
        return z

model = MatrixMul()
poptorch_model = poptorch.inferenceModel(model)

z = poptorch_model(x, w)
print(z)
```
