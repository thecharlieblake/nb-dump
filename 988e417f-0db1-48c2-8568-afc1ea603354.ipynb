```python
import torch
import poptorch

x = torch.randn(2**8, 2**10)
w = torch.randn(2**10, 2**12)

class MatrixMultiplication(torch.nn.Module):
    def forward(self, x, w):
        y = x @ w
        z = y.std()
        return z

model = MatrixMultiplication()
poptorch_model = poptorch.inferenceModel(model)

z = poptorch_model(x, w)
print(z)
```
