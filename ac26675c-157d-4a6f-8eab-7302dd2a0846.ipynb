```python
import torch
import poptorch

class SimpleModel(torch.nn.Module):
    def __init__(self):
        super().__init__()

    def forward(self, x, w):
        y = x @ w * 2**-5
        z = y.std()
        return z

# Create the model
model = SimpleModel()

# Wrap the model for inference on IPU
poptorch_model = poptorch.inferenceModel(model)

# Input tensors
x = torch.randn(2**8, 2**10)
w = torch.randn(2**10, 2**12)

# Run the model on IPU and print the result
z = poptorch_model(x, w)
print(z)
```
