```python
import torch
import poptorch

class MatrixMultiplication(torch.nn.Module):
    def __init__(self, w):
        super().__init__()
        self.w = torch.nn.Parameter(w, requires_grad=False)

    def forward(self, x):
        y = self.w @ x
        return y.std()

# Input tensors
x = torch.randn(2**10, 2**12)
w = torch.randn(2**12, 2**17)

# Create the model
model = MatrixMultiplication(w)

# Create the PopTorch inference model
poptorch_model = poptorch.inferenceModel(model)

# Run the model on IPU and print the result
z = poptorch_model(x)
print(z)
```
