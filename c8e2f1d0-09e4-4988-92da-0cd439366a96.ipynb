```python
import torch
import poptorch

# Create input tensors
x = torch.randn(2**10, 2**12)
w = torch.randn(2**12, 2**17)

# Define a simple model for matrix multiplication and standard deviation
class SimpleModel(torch.nn.Module):
    def forward(self, x, w):
        y = x @ w
        z = y.std()
        return z

# Instantiate the model
model = SimpleModel()

# Wrap the model in a PopTorch inference wrapper
poptorch_model = poptorch.inferenceModel(model)

# Run the model on the IPU
z = poptorch_model(x, w)

print(z)
```
