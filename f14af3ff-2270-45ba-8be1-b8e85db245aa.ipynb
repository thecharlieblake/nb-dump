```python
import torch
import poptorch

x = torch.randn(2**10, 2**12)
w = torch.randn(2**12, 2**14)

# Wrap the matrix multiplication in a model
class MatMulModel(torch.nn.Module):
    def __init__(self, w):
        super().__init__()
        self.w = torch.nn.Parameter(w)

    def forward(self, x):
        y = x @ self.w
        z = y.std()
        return z

model = MatMulModel(w)

# Create an inference model for running on the IPU
inference_model = poptorch.inferenceModel(model)

# Run the model on the IPU and print the result
result = inference_model(x)
print(result)
```
